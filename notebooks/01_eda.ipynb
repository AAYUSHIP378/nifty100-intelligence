# Nifty 100 Exploratory Data Analysis
Quick EDA on the star-schema warehouse CSV files.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

ROOT = Path('..').resolve()
WH = ROOT / 'data' / 'warehouse'
company = pd.read_csv(WH / 'dim_company.csv')
pl = pd.read_csv(WH / 'fact_profit_loss.csv')
scores = pd.read_csv(WH / 'fact_ml_scores.csv')
print('Companies:', len(company))
print('P&L rows:', len(pl))
print('ML score rows:', len(scores))
```

```python
latest = pl[pl['year_label'] != 'TTM'].sort_values(['symbol', 'fiscal_year']).groupby('symbol').tail(1)
sector_health = latest.merge(company[['symbol', 'sector']], on='symbol').groupby('sector')['opm_pct'].mean().sort_values(ascending=False)
sector_health.head(10)
```

```python
annual = pl[pl['year_label'] != 'TTM']
trend = annual.groupby('fiscal_year')['sales'].sum().reset_index()
sns.lineplot(data=trend, x='fiscal_year', y='sales')
plt.title('Total Nifty 100 Revenue Trend')
plt.show()
```
